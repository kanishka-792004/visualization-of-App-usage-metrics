```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```
# Read the dataset
df = pd.read_csv("screentime_analysis.csv")
```

```
print(df.head(1))
```

```
print(df.tail(1))
```

```
# Check missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```
# Drop duplicates
df.drop_duplicates(inplace=True)
print("\nDuplicate Values Removed. New Shape:", df.shape)
```

```
# Convert categorical columns (like "App" names) to numeric using Label Encoding
le = LabelEncoder()
df["App"] = le.fit_transform(df["App"])

print("\nConverted Categorical Columns:\n", df.head())
```

```
# Drop unnecessary columns (e.g., 'Date') and define features & target
features = df.drop(["Usage (minutes)", "Date"], axis=1, errors="ignore")
target = (df["Usage (minutes)"] > df["Usage (minutes)"].median()).astype(int)  # 1 for high usage, 0 for low

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Scale data for better performance in SVM & KNN
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("\nFeature & Target Defined Successfully.")
```

```
import time  # Import time module

# Define models
models = {
    "Support Vector Machine": SVC(kernel="linear"),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Logistic Regression": LogisticRegression()
}

# Train and evaluate each model
for name, model in models.items():
    start_time = time.time()  # Start timing before training

    # Train the model
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Accuracy & Training Time
    accuracy = accuracy_score(y_test, y_pred)
    train_time = time.time() - start_time  # Calculate training time

    print(f"\n{name} Accuracy: {accuracy:.2f}")
    print(f"Training Time: {train_time:.4f} seconds")
```

```
# Screen Time Distribution
plt.figure(figsize=(4,2))
sns.histplot(df["Usage (minutes)"], bins=30, kde=True)
plt.title("Screen Time Distribution")
plt.xlabel("Usage (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```
# Lists to store accuracy results
model_names = []
accuracies = []
for name, model in models.items():
    model.fit(X_train, y_train)  # Train model
    y_pred = model.predict(X_test)  # Make predictions
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy

    # Store results
    model_names.append(name)
    accuracies.append(accuracy)

# Ensure results_df is correctly created
results_df = pd.DataFrame({
    "Model": model_names,
    "Accuracy": accuracies
})

# Print results_df to verify data
print(results_df)

# Plot Accuracy Comparison
plt.figure(figsize=(5,5))
sns.barplot(x="Model", y="Accuracy", data=results_df, palette="coolwarm")
plt.title("Accuracy Comparison of SVM, KNN, and Logistic Regression")
plt.ylabel("Accuracy Score")
plt.ylim(0, 1)
plt.show()
```
